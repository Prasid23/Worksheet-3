```python
#Workshop-3

#Load the provided dataset and import in pandas DataFrame.

# Import the pandas library to work with data
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("/content/drive/MyDrive/DATASET/Copy of Titanic-Dataset.csv")

# Show basic information about the data (columns, data types, and missing values)
df.info()
print("\n")

# Find and print the names of columns with text data (object type)
object_columns = df.select_dtypes(include=['object']).columns
print("Columns with text data:", object_columns)
print("\n")

# Print all the unique values in each column
for col in df.columns:
    print(f"Unique values in {col}:", df[col].unique(), "\n")
print("\n")

# Count and show the number of missing values in each column
for col in df.columns:
    print(f"Missing values in {col}:", df[col].isnull().sum(), "\n")

# Select only the number columns (exclude text columns)
df_numeric = df.select_dtypes(exclude=['object'])
print("Text columns are removed, only numbers remain.")
print("\n")

# Save the numeric data to a new CSV file
numeric_csv_file = "bank_numeric_data.csv"
df_numeric.to_csv(numeric_csv_file, index=False)
print(f"Saved numeric data to: {numeric_csv_file}")
print("\n")

# Load the saved numeric data file and show a summary
df_numeric_loaded = pd.read_csv(numeric_csv_file)
print("Summary of numeric data:")
print(df_numeric_loaded.describe())

```
